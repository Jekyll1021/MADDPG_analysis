```python
%matplotlib inline

import pickle
import numpy as np
import matplotlib.pyplot as plt
import os
plt.style.use('ggplot')
```

```python
def plot_folder(scenario_name):
    plt.figure()
    for comb in os.listdir(scenario_name):
        if os.path.isdir(scenario_name+"/"+comb):
            pickle_off = open(scenario_name+"/"+comb+"/rewards.pkl","rb")
            emp = pickle.load(pickle_off)
            plt.plot(range(len(emp)), emp, label=comb)
    plt.xlabel("epochs")
    plt.ylabel("mean rewards")
    plt.legend()
    plt.title(scenario_name)
```

```python
plot_folder("adv_navigate")
```

```python
plot_folder("coop_navigate")
```
